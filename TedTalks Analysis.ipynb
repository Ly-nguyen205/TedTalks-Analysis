```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel ("TEDTALK x D'CHALLENGER.xlsx")
```

```python
df.info()
```

```python
df1=df.dropna()
```

```python
df1.head(20)
```

```python
##Convert non-latin characters in TITLE and AUTHOR columns to latin
def conv(s):
    return s.encode('latin1','ignore').decode('utf8','ignore')


df1['TITLE']=df1['TITLE'].apply(lambda x: conv(x))
df1['AUTHOR']=df1['AUTHOR'].apply(lambda x: conv(x))
```

```python
#Export for visualisation in Tableau
df1.to_excel('TEDTALK ANALYSIS.xlsx')
```
